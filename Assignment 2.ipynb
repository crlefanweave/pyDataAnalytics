```python
'''Author: Cody LeFan-Weaver
Assignment 2 CS321'''

import numpy as np
import csv
import random
import matplotlib.pyplot as plt
from collections import Counter

grads = []
counter = 0
mean = 0
median = 0
minGPA = 0
maxGPA = 0
q1 = 0
q3 = 0
stdv = 0
outliers = []
scienceMajors = 0
artMajors = 0
civicsMajors = 0
psychMajors = 0

'''preProcessing is something I am doing to fill in blank information for
students in the fields of county, honors, and gpa (which is something I made up).
This eliminates all empty fields. The data for each student is ultimately stored as an array 
of maps, which is held within an array. The structure is a two-dimensional array.'''
def preProcessing():
    global grads
    with open('grads.csv', 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        list = []
        for row in reader:
            if (row['Honors'] == ''):
                row['Honors'] = 'None'
                setGPA(row)
            else:
                row['GPA'] = 4.00
            if (row['Indiana County'] == ''):
                row['Indiana County'] = 'None'
            if (row['State'] == ''):
                row['State'] = 'FOREIGN'
            list.append(row)
    grads = list
                
            
                
def setGPA(stud):
    x = round(random.normalvariate(3.00, 0.33), 2)
    if (x > 3.99):
        x = 3.99
    stud['GPA'] = x

'''The main function will call on other functions to execute the program.'''
def part2():
    global grads

    findMinMax(grads)
    count(grads)
    meanMedian(grads)
    quartiles(grads)
    standDev(grads)
    findOutliers(grads)
    
def generateList():
    with open('grads.csv', 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        list = []
        for row in reader:
            list.append(row)
    return list

'''Beginning here are the functions to print the values we want from the data. This will be qualitative
data focused around the GPA that we have generated.'''
def findMinMax(list):
    global minGPA
    global maxGPA
    n = []
    for x in list:
        n.append(x['GPA'])
    minGPA = np.min(n)
    maxGPA = np.max(n)
    print("The min GPA is: %(min)f, the max GPA is:  %(max)f" % {'min':minGPA, 'max':maxGPA})

def count(list):
    global counter
    for x in list:
        counter+=1
    print("The number of graduate students is %(num)d" % {'num':counter})
    
def meanMedian(list):
    n = []
    global mean
    global median
    for x in list:
        n.append(x['GPA'])
    mean = round(np.mean(n), 2)
    median = round(np.median(n), 2)
    print("The mean GPA of graduates is %(num)f" % {'num':mean})
    print("The median GPA of graduates is %(num)f" % {'num':median})
    
def quartiles(list):
    n = []
    global q1
    global q3
    global counter
    for x in list:
        n.append(x['GPA'])
    n.sort()
    if (counter%4==0):
        q1 = n[counter//4]
        q3 = n[3*counter//4]
    else:
        q1 = (n[int(np.floor(counter/4))] + n[int(np.ceil(3*counter/4))])/2
        q3 = (n[int(np.floor(3*counter/4))] + n[int(np.ceil(3*counter/4))])/2
    print("Q1 is %(num)f" % {'num':q1})
    print("Q3 is %(num)f" % {'num':q3})
    
def standDev(list):
    n = []
    global stdv
    for x in list:
        n.append(x['GPA'])
    stdv = np.std(n)
    print("The standard deviation for this dataset is: %(num)f" % {'num':stdv})
    
def findOutliers(list):
    global mean
    global stdv
    global outliers
    print("The following are outliers in the dataset for GPA: ")
    for x in list:
        if ((x['GPA'] > mean+3*stdv) or (x['GPA'] < mean-3*stdv)):
            outliers.append(x)
            print(x['GPA'])

'''Beginning here will be the functions for part 3, in which I will examine
the number of students graduating from arts and sciences as compared to the total number of students.'''

def findDegreeType():
    global grads
    global counter
    global scienceMajors
    global artMajors
    global civicsMajors
    global psychMajors
    for x in grads:
        if 'SCIENCE' in x['Degree']:
            scienceMajors+=1
        elif 'ART' in x['Degree']:
            artMajors+=1
        elif 'MUS' in x['Degree']:
            artMajors+=1
        elif 'URB' in x['Degree']:
            civicsMajors+=1
        elif 'ARCH' in x['Degree']:
            civicsMajors+=1
        elif 'SOCIAL' in x['Degree']:
            psychMajors+=1
        elif 'PSYC' in x['Degree']:
            psychMajors+=1

'''This will output the percentage of all students who graduated with a degree in science or arts.'''
def outputDegreeStats():
    global counter
    global scienceMajors
    global artMajors
    global civicsMajors
    global psychMajors
    print("There are %(s)d science majors and %(a)d art majors." % {'s':scienceMajors, 'a':artMajors})
    print("There are %(c)d civics majors and %(p)d psych majors." % {'c':civicsMajors, 'p':psychMajors})
    print("Science majors make up %(num)f percent of all graduates." % {'num':100*scienceMajors/counter})
    print("Art majors make up %(num)f percent of all graduates." % {'num':100*artMajors/counter})

'''This creates a pie chart to show the relationship between the different types of degrees
that students graduate from Ball State with.'''
def visualizeDegreeStats():
    global scienceMajors
    global artMajors
    global civicsMajors
    global psychMajors
    global counter
    
    other = counter - (scienceMajors + artMajors + civicsMajors + psychMajors)
    x = [scienceMajors, artMajors, civicsMajors, psychMajors, other]
    
    plt.pie(x, explode=None, labels=['Science', 'Art', 'Civics', 'Psychology', 'Other'],
    colors=('b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'),
    autopct=None, pctdistance=0.6, shadow=False,
    labeldistance=1.1, startangle=None, radius=None,
    counterclock=True, wedgeprops=None, textprops=None,
    center = (0, 0), frame = False )
    plt.show()

'''Shows a histogram to visualize the spread of GPAs among those who graduate from Ball State.'''
def visualizeGPASpread():
    global grads
    gpa = []
    for x in grads:
        gpa.append(x['GPA'])
    
    plt.hist(gpa)
    plt.show()
    
'''This will help to visualize the number of students who graduate from Ball State based on
where they have come from and live.'''
def visualizeHome():
    global grads
    loc = []
    data = []
    for x in grads:
        if (x['State'] == 'INDIANA'):
            loc.append('Indiana')
        elif (x['State'] == 'FOREIGN'):
            loc.append('Other Country')
        else:
            loc.append('Other State')
    labSet = set(loc)
    lab = list(labSet)
    count = Counter(loc)
    
    plt.pie(list(count.values()), explode=None, labels=lab,
    colors=('b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'),
    autopct=None, pctdistance=0.6, shadow=False,
    labeldistance=1.1, startangle=None, radius=None,
    counterclock=True, wedgeprops=None, textprops=None,
    center = (0, 0), frame = False )
    plt.show()
```
